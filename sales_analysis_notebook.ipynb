# Sales Data Analysis Project
This notebook generates synthetic sales data, performs analysis, and visualizes the results using Python.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate synthetic sales data
np.random.seed(42)
data = pd.DataFrame({
    'Product': np.random.choice(['A', 'B', 'C', 'D'], size=200),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], size=200),
    'Units Sold': np.random.randint(5, 100, size=200),
    'Unit Price': np.random.uniform(10, 100, size=200),
    'Date': pd.date_range(start='2023-01-01', periods=200, freq='D')
})

# Calculate total sales
data['Total Sales'] = data['Units Sold'] * data['Unit Price']
data.head()
```

## Summary Statistics

```python
summary = data.describe()
summary
```

## Total Sales per Product

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Product', y='Total Sales', data=data, estimator=sum, ci=None)
plt.title('Total Sales per Product')
plt.tight_layout()
plt.show()
```

## Units Sold by Region

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Region', y='Units Sold', data=data)
plt.title('Units Sold Distribution by Region')
plt.tight_layout()
plt.show()
```

## Distribution of Unit Prices

```python
plt.figure(figsize=(8, 5))
sns.histplot(data['Unit Price'], bins=15, kde=True)
plt.title('Distribution of Unit Prices')
plt.tight_layout()
plt.show()
```
